```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import Input, layers, models, datasets, utils, losses, optimizers 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.utils import to_categorical


from tensorflow.keras.layers import Dropout, Flatten, Dense
from tensorflow.keras.layers import Conv2D, MaxPooling2D
#from tensorflow.keras.layers.normalization import BatchNormalization
from tensorflow.keras.layers import LeakyReLU
```

```python
df = pd.read_csv('Breast_GSE45827.csv')
```

```python
df.head()
```

```python
new_df = df[df['type'] != 'cell_line' ]
new_df = new_df[new_df['type'] != 'normal']
new_df.shape   
```

```python
new_df['type'].value_counts()
```

```python
X = new_df.drop(['samples', 'type'],axis = 1)


from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(new_df['type']) 


```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
#Standardization of data
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(256, input_shape = (54675,), activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(4,activation = 'softmax'))

```

```python
model.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = 'SGD',
              metrics = ['accuracy']) 
```

```python
model.summary()
```

```python
#from tensorflow.keras.utils import plot_model
#plot_model(model = model, show_shapes=True)   
```

```python
model.fit(X_train_scaled, y_train, epochs = 6)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn import metrics 
```

```python
loss, acc = model.evaluate(X_test_scaled, y_test)
print(loss)
print(acc)
```
